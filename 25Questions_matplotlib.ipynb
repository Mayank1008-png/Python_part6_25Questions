```python
#Ans12
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y1=[4,7,15,8,12]
y2=[5,7,3,23,12]
y3=[7,21,14,17,8]
y4=[11,6,15,1.5,5]
fig , ax = plt.subplots(2,2,figsize=(10,6))
ax[0,0].plot(x,y1,marker='+',linestyle='--',color='green',linewidth=2)
ax[0,0].set_title('Line Graph 1',fontsize=15)

ax[0,1].bar(x,y2,color=['teal','violet','purple','coral','red'])
ax[0,1].set_title('Bar Graph',fontsize=15)

ax[1,0].scatter(x,y3,color=['teal','violet','purple','coral','red'])
ax[1,0].set_title('Scatter Graph',fontsize=15)

ax[1,1].pie(y4,labels=x,colors=['teal','violet','purple','coral','red'],autopct='%1.1f%%')
ax[1,1].set_title('Pie Graph',fontsize=15)
plt.tight_layout()
plt.suptitle('4 Graphs')
plt.tight_layout()
plt.show()
```

```python
#Ans13
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,200)
#print(x)
y1=np.sin(x)
y2=np.cos(x)
fig , ax = plt.subplots(2,1,sharex=True,figsize=(10,5))

plt.ylim(-1,1)
ax[0].plot(x,y1)
ax[0].set_title('SINNE WAVE')
ax[0].set_xlabel("X Axis")
ax[0].set_ylabel("Amplitude")

ax[1].plot(x,y2)
ax[1].set_title('Cosine WAVE')
ax[1].set_xlabel("X Axis")
ax[1].set_ylabel("Amplitude")

plt.tight_layout()

plt.show()
```

```python
#Ans17
import matplotlib.pyplot as plt

math_marks = [78, 85, 92, 88, 76, 81, 95, 89, 84, 77]
science_marks = [72, 80, 91, 85, 79, 83, 87, 90, 82, 75]
english_marks = [70, 78, 88, 84, 73, 80, 86, 82, 79, 76]
data1=[math_marks,science_marks,english_marks]
box=plt.boxplot(data1,patch_artist=True,labels=['Maths','Science','English'])
colors=['red','yellow','green']
for patch,color in zip(box['boxes'],colors):
    patch.set_facecolor(color)
plt.title('Marks of Students')
plt.grid(True)
plt.ylabel('Marks')
plt.show()
```

```python
#Ans19
import matplotlib.pyplot as plt
math = [78, 85, 92, 88, 76, 81, 95, 89, 84, 77]
science = [72, 80, 91, 85, 79, 83, 87, 90, 82, 75]
english = [70, 78, 88, 84, 73, 80, 86, 82, 79, 76]
data = [math, science, english]
plt.violinplot(data, showmeans=True, showmedians=True)
plt.xticks([1, 2, 3], ['Math', 'Science', 'English'])
plt.title("Violin Plot of Student Marks")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
#Ans20
import matplotlib.pyplot as plt
import numpy as np
students = ['Mayank', 'Dev', 'Nishu', 'Somveer', 'Arun']
math_marks = [80, 85, 78, 90, 88]
science_marks = [75, 80, 70, 85, 82]
english_marks = [72, 78, 75, 80, 77]

# Bar positions
x = np.arange(len(students))

# Plot base (math)
plt.bar(x, math_marks, label='Math', color='skyblue')

# Stack science on top of math
plt.bar(x, science_marks, bottom=math_marks, label='Science', color='lightgreen')

# Calculate bottom for english
bottom_english = np.array(math_marks) + np.array(science_marks)
plt.bar(x, english_marks, bottom=bottom_english, label='English', color='lightcoral')


plt.xticks(x, students)
plt.xlabel("Students")
plt.ylabel("Total Marks")
plt.title("Stacked Bar Chart of Student Marks")
plt.legend()


plt.show()
```

```python
#ans23
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Fill area under curve
plt.plot(x, y, label='sin(x)', color='blue')
plt.fill_between(x, y, color='skyblue', alpha=0.4)

# Labels
plt.title("Area Under Sine Curve")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Ans24
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(1, 6)
y = np.array([20, 35, 30, 35, 27])
error = np.array([2, 3, 4, 1, 2])  # example error values

# Line plot with error bars
plt.errorbar(x, y, yerr=error, fmt='o-', capsize=5, color='darkorange', label='Scores with Error')

# Labels
plt.title("Line Plot with Error Bars")
plt.xlabel("Test Number")
plt.ylabel("Score")
plt.legend()
plt.grid(True)
plt.show()
```
